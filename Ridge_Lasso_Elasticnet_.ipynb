```python
X = df.drop('mpg', axis =1)
```

```python
y=df['mpg']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
regression_nodel = LinearRegression()
regression_nodel
```

```python
regression_nodel.fit(X_train, y_train)
```

```python
for i, col_name in enumerate(X_train.columns):
  print(f"The cofficients for {col_name} is {regression_nodel.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred_linear = regression_nodel.predict(X_test)
```

```python
r2_score(y_test,y_pred_linear)
```

```python
#Ridge regression

from sklearn.linear_model import Ridge
```

```python
ridge_regressor_model = Ridge(alpha = 100)
ridge_regressor_model.fit(X_train, y_train)
for i, col_name in enumerate(X_train.columns):
  print(f"The cofficient for {col_name} is {ridge_regressor_model.coef_[i]}")
```

```python

y_pred = ridge_regressor_model.predict(X_test)
r2_score(y_test, y_pred)
```

```python
#Lasso
from sklearn.linear_model import Lasso

lasso_regressor_model = Lasso(alpha = 0.5)
lasso_regressor_model.fit(X_train, y_train)
for i, col_name in enumerate(X_train.columns):
  print(f"The cofficient for {col_name} is {lasso_regressor_model.coef_[i]}")

#By o/p we can say that even if we remove features cylinders, accleration still we get Accuracy or r2 score 83 as earlier
```

```python
y_pred = lasso_regressor_model.predict(X_test)
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import ElasticNet

elastic_regression_model = ElasticNet(alpha=0.8, l1_ratio=0.4)
elastic_regression_model.fit(X_train, y_train)
for i, col_name in enumerate(X_train.columns):
    print(f"The coefficient for {col_name} is {elastic_regression_model.coef_[i]}")
```

```python
from sklearn.metrics import r2_score
y_pred = elastic_regression_model.predict(X_test)
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv = 5)
lassocv.fit(X_train, y_train)
y_pred = lassocv.predict(X_test)
r2_score(y_test, y_pred)
```
