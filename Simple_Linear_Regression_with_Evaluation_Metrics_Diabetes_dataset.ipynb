```python
data = pd.DataFrame(diabetes.data, columns= diabetes.feature_names)
```

```python
#Combining Target column in created Dataframe  so that we able to understand effect of other conditions on disease progression  one year after baseline
data['target'] = diabetes.target
```

```python
data
```

```python
#Data preparation
# Divide data in X and y
```

```python
#Simple Linear Regression

X = data[['bmi']]
y = data['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state =111)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#Scaling
#Model Training

from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
#Slope or m or beta1,  theat1 or bais(b) in Deep Learining

model.coef_
```

```python
# constant or c or betanot , theatnot or weight(w) in Deep Learining

model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
X_test
```

```python
y_pred
```

```python
data2 = pd.DataFrame(X_test,columns=y_pred)
```

```python
data2
```

```python
#Visualize the result

plt.scatter(X_test, y_test, color = 'black', label = "Actual data")
plt.plot(X_test, y_pred, color = 'red',linewidth = 3, label = "Linear Regression Line")
plt.xlabel("BMI")
plt.ylabel("One Year progression-target")
plt.title("My First Linear  Regression Model")
plt.legend()
plt.show('True')
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), 'r')
```
